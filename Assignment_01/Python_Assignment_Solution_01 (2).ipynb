# Python Assignment Notes (Theory Questions 1–14)


## 1. What is Python, and why is it popular?

### Answer

a) Python is a high-level, interpreted, general-purpose programming language.

b) It was created by Guido van Rossum in 1991.

c) It is easy to read and write because of its simple syntax.

d) It supports Object-Oriented, Functional, and Procedural programming.

e) It has a large standard library and many third-party libraries.

f) It is open-source and free to use.

g) It is widely used in Web Development, Data Science, Machine Learning, Artificial Intelligence, Automation, and Cyber Security.



## 2. What is an interpreter in Python?

### Answer

a) An interpreter executes Python code line by line.

b) It converts Python statements into machine instructions at runtime.

c) It makes debugging easier because errors are reported immediately.



## 3. What are pre-defined keywords in Python?

### Answer

a) Keywords are reserved words with predefined meanings.

b) They cannot be used as variable names.

c) Examples: if, else, for, while, def, class, return, True, False, None.



## 4. Can keywords be used as variable names?

### Answer

a) No. Keywords cannot be used as variable names.

b) Using a keyword as a variable name raises a SyntaxError.



## 5. What is mutability in Python?

### Answer

a) Mutability means an object can be changed after it is created.

b) Lists, dictionaries, and sets are mutable.

c) Strings and tuples are immutable.



## 6. Why are lists mutable, but tuples are immutable?

### Answer

a) Lists are designed to allow insertion, deletion, and modification of elements.

b) Tuples are fixed after creation, making them faster and safer for constant data.



## 7. What is the difference between '==' and 'is' operators in Python?

### Answer

a) '==' compares values.

b) 'is' compares object identity (memory location).



## 8. What are logical operators in Python?

### Answer

a) Python provides three logical operators:

b) and – True if both conditions are True.

c) or – True if at least one condition is True.

d) not – Reverses the Boolean value.



## 9. What is type casting in Python?

### Answer

a) Type casting is converting one data type into another.

b) Examples: int(), float(), str(), bool().



## 10. What is the difference between implicit and explicit type casting?

### Answer

a) Implicit type casting is performed automatically by Python.

b) Explicit type casting is performed manually using conversion functions like int() and float().



## 11. What is the purpose of conditional statements in Python?

### Answer

a) Conditional statements make decisions based on conditions.

b) Python provides if, elif, and else statements.



## 12. How does the elif statement work?

### Answer

a) elif means 'else if'.

b) It checks another condition when the previous if condition is False.

c) It avoids writing multiple nested if statements.



## 13. What is the difference between for and while loops?

### Answer

a) for loop is used when the number of iterations is known.

b) while loop is used when the number of iterations is unknown and depends on a condition.



## 14. Describe a scenario where a while loop is more suitable than a for loop.

### Answer

a) A while loop is suitable when the number of repetitions is not known in advance.

b) Examples: ATM PIN verification, login systems, menu-driven programs, and user input validation.
