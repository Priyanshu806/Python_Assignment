## 6. Mutable vs Immutable arguments

### Problem Statement
6. Mutable vs Immutable arguments

````
def modify_list(l):
    l.append(4)

def modify_num(n):
    n+=5

lst=[1,2,3]
num=10
modify_list(lst)
modify_num(num)
print(lst)
print(num)
````

### Sample Output
```
[1, 2, 3, 4]
10
```

### Explanation
a) Lists change because they are mutable.

b) Integers remain unchanged because they are immutable.

## 14. Reverse a string using while

### Problem Statement
14. Reverse a string using while

````
s=input("Enter string: ")
rev=""
i=len(s)-1
while i>=0:
    rev+=s[i]
    i-=1
print(rev)
````

### Sample Output
```
Reverse of input
```

### Explanation
a) Traverses the string from end to start.



## 15. Factorial using while

### Problem Statement
15. Factorial using while

````
n=int(input("Enter number: "))
f=1
while n>1:
    f*=n
    n-=1
print(f)
````

### Sample Output
```
Depends on input
```

### Explanation
a) Calculates factorial using a while loop.
