## 2. Abstract Shape

### Problem Statement
2. Abstract Shape

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        return 3.14*self.r*self.r

class Rectangle(Shape):
    def __init__(self,l,b):
        self.l=l
        self.b=b
    def area(self):
        return self.l*self.b

print(Circle(5).area())
print(Rectangle(4,6).area())
```

### Explanation
- Implements abstraction.
- Demonstrates the required OOP concept.

## 3. Multi-level Inheritance

### Problem Statement
3. Multi-level Inheritance

```python
class Vehicle:
    def __init__(self,t):
        self.type=t

class Car(Vehicle):
    pass

class ElectricCar(Car):
    def __init__(self,t,battery):
        super().__init__(t)
        self.battery=battery

e=ElectricCar("SUV","75 kWh")
print(e.type,e.battery)
```

## 4. Bird Polymorphism

### Problem Statement
4. Bird Polymorphism

```python
class Bird:
    def fly(self):
        print("Bird flies")

class Sparrow(Bird):
    def fly(self):
        print("Sparrow flies")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

for b in [Sparrow(),Penguin()]:
    b.fly()
```

### Explanation
- Runtime polymorphism.
- Demonstrates the required OOP concept.

## 5. BankAccount Encapsulation

### Problem Statement
5. BankAccount Encapsulation

```python
class BankAccount:
    def __init__(self):
        self.__balance=0
    def deposit(self,a):
        self.__balance+=a
    def withdraw(self,a):
        self.__balance-=a
    def check(self):
        return self.__balance

b=BankAccount()
b.deposit(500)
b.withdraw(100)
print(b.check())
```

### Explanation
- Private attribute example.
- Demonstrates the required OOP concept.

## 7. Class & Static Method

### Problem Statement
7. Class & Static Method

```python
class MathOperations:
    @classmethod
    def add_numbers(cls,a,b):
        return a+b
    @staticmethod
    def subtract_numbers(a,b):
        return a-b

print(MathOperations.add_numbers(5,3))
print(MathOperations.subtract_numbers(5,3))
```

## 10. Operator Overloading

### Problem Statement
10. Operator Overloading

```python
class Vector:
    def __init__(self,x,y):
        self.x=x; self.y=y
    def __add__(self,o):
        return Vector(self.x+o.x,self.y+o.y)
    def __str__(self):
        return f"({self.x},{self.y})"
print(Vector(1,2)+Vector(3,4))
```

### Explanation
- Overloads + operator.
- Demonstrates the required OOP concept.

## 14. Employee & Manager

```python
class Employee:
    def calculate_salary(self,h,r):
        return h*r
class Manager(Employee):
    def calculate_salary(self,h,r):
        return super().calculate_salary(h,r)+5000
print(Manager().calculate_salary(20,500))
```

## 18. House & Mansion

### Problem Statement
18. House & Mansion

```python
class House:
    def __init__(self,address,price):
        self.address=address; self.price=price
class Mansion(House):
    def __init__(self,address,price,rooms):
        super().__init__(address,price)
        self.rooms=rooms
m=Mansion("Delhi",5000000,12)
print(m.address,m.rooms)
```

### Explanation
- Inheritance example.
- Demonstrates the required OOP concept.
