## 4. Difference between abstraction and encapsulation?

### Answer

a) Abstraction hides implementation details.

b) Encapsulation bundles data and methods together and restricts direct access.



## 5. What are dunder methods in Python?

### Answer

a) Dunder (double underscore) methods are special methods.

b) Examples: __init__, __str__, __len__, __repr__.



```python
class Demo:
    def __str__(self):
        return "Demo Object"
print(Demo())
```

## 6. Explain inheritance in OOP.

### Answer

a) Inheritance allows one class to acquire properties of another.

b) It promotes code reuse.



```python
class Animal:
    def sound(self):
        print("Sound")
class Dog(Animal):
    pass
Dog().sound()
```
