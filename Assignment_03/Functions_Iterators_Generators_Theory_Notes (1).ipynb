## 1. What is the difference between a function and a method in Python?

### Answer

a) A function is defined independently using def.

b) A method is a function associated with an object or class.

c) Example: len(text) is a function, text.upper() is a method.



```python
text='python'
print(len(text))      # Function
print(text.upper())  # Method
```

## 2. Explain the concept of function arguments and parameters in Python.

### Answer

a) Parameters are variables in the function definition.

b) Arguments are values passed during the function call.

c) Example: def add(a,b): return a+b; add(2,3).



```python
def add(a,b):
    return a+b

print(add(2,3))
```

## 5. What are iterators in Python and how do they differ from iterables?

### Answer

a) An iterable is an object you can loop over.

b) An iterator produces one element at a time using next().

c) Example: list is iterable; iter(list) returns an iterator.



## 6. Explain the concept of generators and how they are defined.

### Answer

a) Generators produce values lazily.

b) Defined using yield instead of return.

c) They save memory for large datasets.



```python
def gen():
    for i in range(3):
        yield i

for x in gen():
    print(x)
```

## 7. What are the advantages of using generators over regular functions?

### Answer

a) Memory efficient.

b) Generate values on demand.

c) Useful for processing large files or streams.



## 8. What is a lambda function in Python and when is it typically used?

### Answer

a) A lambda is an anonymous one-line function.

b) Commonly used with map(), filter(), and sorted().

c) Example: lambda x: x*x.



```python
square=lambda x:x*x
print(square(5))
```

## 9. Explain the purpose and usage of the map() function in Python.

### Answer

a) map() applies a function to every element of an iterable.

b) Returns a map object.

c) Example: map(lambda x:x*2,[1,2,3]).



```python
nums=[1,2,3]
print(list(map(lambda x:x*2,nums)))
```

## 10. What is the difference between map(), reduce(), and filter() functions in Python?

### Answer

a) map() transforms every element.

b) filter() keeps elements matching a condition.

c) reduce() combines all elements into one value (from functools).



```python
from functools import reduce
nums=[47,11,42,13]
print(list(map(lambda x:x*2,nums)))
print(list(filter(lambda x:x>20,nums)))
print(reduce(lambda a,b:a+b,nums))
```
