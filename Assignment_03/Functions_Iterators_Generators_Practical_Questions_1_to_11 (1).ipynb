## 1. Sum of all even numbers in a list

### Problem Statement
1. Sum of all even numbers in a list

````
def sum_even(nums):
    return sum(x for x in nums if x % 2 == 0)

nums=[1,2,3,4,5,6]
print(sum_even(nums))
````

### Sample Output
```
12
```

### Explanation
a) Filters even numbers and returns their sum.

## 3. Squares of all numbers

### Problem Statement
3. Squares of all numbers

````
def square_list(nums):
    return [x*x for x in nums]

print(square_list([1,2,3,4]))
````

### Sample Output

```
[1, 4, 9, 16]
```

## 4. Check prime number (1-200)

### Problem Statement
4. Check prime number (1-200)

````
def is_prime(n):
    if n<2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True

print(is_prime(197))
````

### Sample Output
```
True
```

### Explanation
a) Checks divisibility up to √n.



## 5. Fibonacci iterator class

### Problem Statement
5. Fibonacci iterator class

````
class Fibonacci:
    def __init__(self,n):
        self.n=n
        self.a,self.b=0,1
        self.count=0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count>=self.n:
            raise StopIteration
        val=self.a
        self.a,self.b=self.b,self.a+self.b
        self.count+=1
        return val

for i in Fibonacci(7):
    print(i,end=' ')
````

### Sample Output
```
0 1 1 2 3 5 8
```

### Explanation
a) Implements iterator protocol.



## 6. Generator powers of 2

### Problem Statement
6. Generator powers of 2

````
def powers(n):
    for i in range(n+1):
        yield 2**i

for x in powers(5):
    print(x)
````

### Sample Output
```
1 2 4 8 16 32
```

### Explanation
a) Uses yield to generate values.



## 7. Read file line by line

### Problem Statement
7. Read file line by line

````
def read_file(file):
    with open(file,'r') as f:
        for line in f:
            yield line.strip()

# Example:
# for line in read_file('sample.txt'):
#     print(line)
````

## 8. Lambda sort tuples

### Problem Statement
8. Lambda sort tuples

````
data=[('A',3),('B',1),('C',2)]
data.sort(key=lambda x:x[1])
print(data)
````

### Sample Output
```
[('B',1),('C',2),('A',3)]
```

### Explanation
a) Sorts by second element.



## 9. Celsius to Fahrenheit using map()

### Problem Statement
9. Celsius to Fahrenheit using map()

````
c=[0,20,30,40]
f=list(map(lambda x:(x*9/5)+32,c))
print(f)
````

### Sample Output

```
[32.0, 68.0, 86.0, 104.0]
```

### Explanation
a) map() converts each value.



## 10. Remove vowels using filter()

### Problem Statement
10. Remove vowels using filter()

````
text='Programming'
result=''.join(filter(lambda ch: ch.lower() not in 'aeiou',text))
print(result)
````

### Sample Output
```
Prgrmmng
```

### Explanation
a) filter() removes vowels.



## 11. Book shop routine

### Problem Statement
11. Book shop routine

````
orders=[
(34587,'Learning Python, Mark Lutz',4,40.95),
(98762,'Programming Python, Mark Lutz',5,56.80),
(77226,'Head First Python, Paul Barry',3,32.95),
(88112,'Einfuhrung in Python3, Bernd Klein',3,24.99)
]

result=[]
for no,title,qty,price in orders:
    total=qty*price
    if total<100:
        total+=10
    result.append((no,total))

print(result)
````

### Sample Output
```
[(34587, 163.8), (98762, 284.0), (77226, 108.85), (88112, 84.97)]
```

### Explanation
a) Adds €10 if order total is below 100.
