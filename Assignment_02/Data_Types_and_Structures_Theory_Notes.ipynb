# Data Types and Structures Questions (Theory Notes)



## 1. What are data structures, and why are they important?

### Answer

a) Data structures organize and store data efficiently.

b) They improve performance.

c) Examples: List, Tuple, Set, Dictionary.



## 2. Explain the difference between mutable and immutable data types with examples.

### Answer

a) Mutable objects can be modified (List, Dictionary, Set).

b) Immutable objects cannot be modified (String, Tuple, Integer).



## 3. What are the main differences between lists and tuples in Python?

### Answer

a) Lists are mutable.

b) Tuples are immutable.

c) Lists use [], Tuples use ().



## 4. Describe how dictionaries store data.

### Answer

a) Dictionaries store data as key-value pairs.

b) Each key is unique.

c) Implemented using hash tables.



## 5. Why might you use a set instead of a list in Python?

### Answer

a) Sets automatically remove duplicates.

b) Membership testing is faster.

c) Useful for unique values.



## 6. What is a string in Python, and how is it different from a list?

### Answer

a) A string is an immutable sequence of characters.

b) A list is mutable and stores different data types.



## 7. How do tuples ensure data integrity in Python?

### Answer

a) Tuples cannot be modified after creation.

b) This prevents accidental changes.



## 8. What is a hash table, and how does it relate to dictionaries in Python?

### Answer

a) A hash table maps keys to values.

b) Python dictionaries use hash tables internally.



## 9. Can lists contain different data types in Python?

### Answer

a) Yes.

b) Lists can store integers, strings, floats, booleans, objects, and other lists.



## 10. Explain why strings are immutable in Python.

### Answer

a) Immutability improves security, performance, and hashing.



## 11. What advantages do dictionaries offer over lists for certain tasks?

### Answer

a) Fast key-based lookup.

b) Unique keys.

c) Efficient insertion and updates.



## 12. Describe a scenario where using a tuple would be preferable over a list.

### Answer

a) When data should never change, such as coordinates or dates.



## 13. How do sets handle duplicate values in Python?

### Answer

a) Duplicate values are automatically removed.



## 14. How does the 'in' keyword work differently for lists and dictionaries?

### Answer

a) For lists, it searches values.

b) For dictionaries, it searches keys.



## 15. Can you modify the elements of a tuple? Explain why or why not.

### Answer

a) No.

b) Tuples are immutable.



## 16. What is a nested dictionary, and give an example of its use case?

### Answer

a) A dictionary inside another dictionary.

b) Example: Student records.



## 17. Describe the time complexity of accessing elements in a dictionary.

### Answer

a) Average case: O(1).

b) Worst case: O(n).



## 18. In what situations are lists preferred over dictionaries?

### Answer

a) When order and indexing matter.

b) When duplicates are allowed.



## 19. Why are dictionaries considered unordered, and how does that affect data retrieval?

### Answer

a) Modern Python preserves insertion order.

b) Retrieval is still by key, not by index.



## 20. Explain the difference between a list and a dictionary in terms of data retrieval.

### Answer

a) Lists use indexes.

b) Dictionaries use keys.
