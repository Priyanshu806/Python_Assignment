## 1. 2D NumPy array and row sums

### Problem Statement
1. 2D NumPy array and row sums

````python
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
print(arr)
print(arr.sum(axis=1))
````

### Sample Output
```
[6 15]
```

### Explanation
- sum(axis=1) computes row sums.
- Demonstrates the required Data Toolkit concept.

## 4. Correlation heatmap (Seaborn)

### Problem Statement

## 13. 3D scatter plot

### Problem Statement
13. 3D scatter plot

````python
import plotly.express as px
import pandas as pd
df=pd.DataFrame({'x':[1,2,3],'y':[2,3,4],'z':[5,6,7]})
fig=px.scatter_3d(df,x='x',y='y',z='z')
fig.show()
````

### Sample Output
```
3D Scatter Plot
```

### Explanation
- Creates interactive 3D scatter plot.
- Demonstrates the required Data Toolkit concept.
